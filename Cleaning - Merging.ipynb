```python
# import pandas
import pandas as pd
```

```python
# Load all cleaned csv data

df_dc = pd.read_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\clean\\Customer DC Inventory.csv')
df_factory = pd.read_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\clean\\Factory POS $.csv')
df_ecomm = pd.read_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\clean\\Total Ecomm POS (Factory $).csv')
df_sales = pd.read_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\clean\\Total Sales.csv')
df_trade = pd.read_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\clean\\Total Trade Spend.csv')
```

```python
# df_sales and df_trade cover slightly different data from the other files, so we'll remove some rows to make them the same
```

```python
# remove Need State 6 from all datasets that have it
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html
# https://saturncloud.io/blog/how-to-remove-rows-with-specific-values-in-pandas-dataframe/#:~:text=Another%20method%20to%20remove%20rows,value%20we%20want%20to%20remove.
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.index.html

df_dc = df_dc.drop(df_dc[df_dc['Need States'] == 'Need State 6'].index)
df_factory = df_factory.drop(df_factory[df_factory['Need States'] == 'Need State 6'].index)
df_ecomm = df_ecomm.drop(df_ecomm[df_ecomm['Need States'] == 'Need State 6'].index)
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html

df_dc = df_dc.reset_index()
df_factory = df_factory.reset_index()
df_ecomm = df_ecomm.reset_index()
```

```python
df_factory
```

```python
df_ecomm
```

```python
# remove the extra weeks covered in 2023 in df_sales and df_trade
# https://www.w3schools.com/python/ref_func_range.asp

for index in range(725, 735):
    df_sales.drop([index], inplace=True)
```

```python
for index in range(725, 745):
    df_trade.drop([index], inplace=True)
```

```python
df_sales
```

```python
df_trade
```

```python
# remove unnecessary columns

df_dc = df_dc.drop(['index', 'Unnamed: 0'], axis=1)
df_factory = df_factory.drop(['index', 'Unnamed: 0'], axis=1)
df_ecomm = df_ecomm.drop(['index', 'Unnamed: 0'], axis=1)
```

```python
df_sales = df_sales.drop(['Unnamed: 0'], axis=1)
df_trade = df_trade.drop(['Unnamed: 0'], axis=1)
```

```python
# merge datasets
# https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html#database-style-dataframe-or-named-series-joining-merging
# https://realpython.com/pandas-merge-join-and-concat/#:~:text=merge()%20for%20combining%20data,DataFrames%20across%20rows%20or%20columns

df = pd.merge(df_dc, df_factory)
```

```python
df
```

```python
df = pd.merge(df, df_ecomm)
```

```python
df
```

```python
df = pd.merge(df, df_sales)
```

```python
df = pd.merge(df, df_trade)
```
