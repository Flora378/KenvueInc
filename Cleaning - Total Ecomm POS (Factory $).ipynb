```python
# import pandas
import pandas as pd
```

```python
# load CSV data and declare row 0 and row 1 as header (multi index)
# https://pandas.pydata.org/docs/user_guide/advanced.html

df = pd.read_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\raw\\Total Ecomm POS (Factory $) - UTSC Lecture.csv', header=[0,1])
```

```python
df
```

```python
df.columns
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.melt.html
df = df.melt(id_vars=[('Unnamed: 1_level_0', 'Fiscal Year'), ('Unnamed: 0_level_0',  'Need State')])
```

```python
df
```

```python
# rename the columns
df = df.rename(columns={df.columns[0]: "Year", df.columns[1]: "Need States", df.columns[3]: "Week", df.columns[4]: "Ecomm POS (Factory $)"})
```

```python
df
```

```python
# remove extra column
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pop.html
df.pop('variable_0')
```

```python
df
```

```python
# week is the incorrect type; it won't sort properly
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dtypes.html
df.dtypes
```

```python
# convert week to int
# https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html#pandas.to_numeric
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html
# https://stackoverflow.com/questions/15891038/change-column-type-in-pandas#:~:text=The%20best%20way%20to%20convert,floating%2Dpoint%20numbers%20as%20appropriate.
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
df.dtypes
```

```python
# convert week# to an actual date for easy plotting
df['Date'] = pd.to_datetime(df["Year"].astype(str) + "-" + df["Week"].astype(str) + "-1", format='%Y-%U-%w')
```

```python
df
```

```python
import seaborn as sns
```

```python
# try a couple graphs
sns.lineplot(data = df, x=df.Date, y='Ecomm POS (Factory $)')
```

```python
sns.lineplot(data = df, x='Week', y='Ecomm POS (Factory $)', hue='Year')
```

## Additional:

```python
# rearrange columns
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reindex.html
# https://stackoverflow.com/questions/13148429/how-to-change-the-order-of-dataframe-columns
df = df.reindex(columns=['Year','Week','Date','Need States','Ecomm POS (Factory $)'])
```

```python
df
```

```python
# sort values so format matches other cleaned csv files
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html
df = df.sort_values(by=['Year', 'Week'])
```

```python
df
```

```python
# reset index numbers
df = df.reset_index(drop=True)
```

```python
df
```

```python
# replace instances of 0 with NaN
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html
# https://stackoverflow.com/questions/34794067/how-to-set-a-cell-to-nan-in-a-pandas-dataframe
df = df.replace(to_replace=0, value=pd.NA)
```

```python
df
```

```python
# remove rows that contain NaN
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html
df = df.dropna(axis=0)
```

```python
df
```

```python
# check how the graphs look now
sns.lineplot(data = df, x=df.Date, y='Ecomm POS (Factory $)')
```

```python
sns.lineplot(data = df, x='Week', y='Ecomm POS (Factory $)', hue='Year')
```

```python
# write the dataframe to a csv file
df.to_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\clean\\Total Ecomm POS (Factory $).csv')
```
