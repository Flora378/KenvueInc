```python
# This is an example of how I would clean and reorganize the data.  First, let's import the library 'Pandas'
import pandas as pd
```

```python
# Loading the CSV data.  Noticed that I decleard row 0 and row 1 as header.  The data actually begins in row 2.
# By doing this, I created a multi index. See documentation: https://pandas.pydata.org/docs/user_guide/advanced.html

df = pd.read_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\raw\\Customer DC Inventory - UTSC Lecture.csv', header=[0,1])
```

```python
# Let's see what it looks like.  Noticed that there are 2 header rows.
df
```

```python
# This is called multi index.  See documentation: https://pandas.pydata.org/docs/user_guide/advanced.html
df.columns
```

```python
# Instead of having numbers in row index, I am going to set customized index levels
df = df.set_index([('Fiscal Year  /  Fiscal Week', 'Need State'), ('Unnamed: 1_level_0', 'Unnamed: 1_level_1')])
```

```python
# Swapping Rows <-> Columns
df = df.T
```

```python
# Let's take a look..getting a bit more organized now.
df
```

```python
# I use a Pandas stack function here to stack all the need states in one column.  
# That way my dataframe is long instead of wide.
# See stack function here: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.stack.html
df = df.stack(0)
```

```python
# Can you see what I did so far?
df
```

```python
# Now I will reset the index so it will go back to numbers as row index
df= df.reset_index()
```

```python
df
```

```python
# Let's rename the columns
df = df.rename_axis(None, axis=1)
df = df.rename(columns={df.columns[0]: "Year", df.columns[1]: "Week", df.columns[2]: "Need States"})
```

```python
df
```

```python
# Here I use the datetime function to convert the week# to an acutual date for easy plotting
df['Date'] = pd.to_datetime(df["Year"].astype(str) + "-" + df["Week"].astype(str) + "-1", format='%Y-%U-%w')
```

```python
df
```

```python
import seaborn as sns
```

```python
sns.lineplot(data = df, x=df.Date, y='Dc Amount')
```

```python
sns.lineplot(data = df, x='Week', y='Dc Amount', hue='Year')
```

## Additional:

```python
# rearrange columns
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reindex.html
# https://stackoverflow.com/questions/13148429/how-to-change-the-order-of-dataframe-columns
df = df.reindex(columns=['Year','Week','Date','Need States','Dc Amount','Store Amt On Hand'])
```

```python
df
```

```python
# replace instances of 0 with NaN
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html
# https://stackoverflow.com/questions/34794067/how-to-set-a-cell-to-nan-in-a-pandas-dataframe
df = df.replace(to_replace=0, value=pd.NA)
```

```python
df
```

```python
# remove rows that contain NaN
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html
df = df.dropna(axis=0)
```

```python
df
```

```python
# check how the graphs look now
sns.lineplot(data = df, x=df.Date, y='Dc Amount')
```

```python
sns.lineplot(data = df, x='Week', y='Dc Amount', hue='Year')
```

```python
# write the dataframe to a csv file
df.to_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\clean\\Customer DC Inventory.csv')
```
