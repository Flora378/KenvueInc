```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\anime\\Desktop\\Kenvue_Final\\kenvue_data\\Total Trade Spend.csv')
```

```python
df['date'] = pd.to_datetime(df["Fiscal Year"].astype(str) + "-" + df["Fiscal Week"].astype(str) + "-1", format='%Y-%U-%w')
```

```python
df
```

```python
df_melt = df.melt(id_vars=["Fiscal Year", "Fiscal Week", "date"], var_name="Need State", value_name="Trade Spend")
```

```python
import seaborn as sns
```

```python
sns.lineplot(data = df_melt, x='date', y='Trade Spend', hue='Need State')
```
